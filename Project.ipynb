```python
import numpy as np
```

```python
import pandas as pd
import xgboost
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df=pd.read_csv('/content/CICIDS2017_sample.csv')
df.head()
```

```python
df
```

```python
df.Label.value_counts()
```

```python
features = df.dtypes[df.dtypes != 'object'].index
features
```

```python
df[features] = df[features].apply(
    lambda x: (x - x.mean()) / (x.std()))
```

```python
df=df.fillna(0)
df
```

```python
features
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
df.iloc[:, -1] = labelencoder.fit_transform(df.iloc[:, -1])
```

```python
df.iloc[:, -1]
```

```python
df.Label.value_counts()
```

```python
minor = df[(df['Label']==6)|(df['Label']==1)|(df['Label']==4)]
major = df.drop(minor.index)
```

```python
X = major.drop(['Label'],axis=1)
y = major.iloc[:, -1].values.reshape(-1,1)
y=np.ravel(y)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=7, random_state=0).fit(X)

```

```python
klabel=kmeans.labels_
major['klabel']=klabel
```

```python
major['klabel'].value_counts()
```

```python
cols = list(major)
cols.insert(100, cols.pop(cols.index('Label')))
major = major.loc[:, cols]
```

```python
major
```

```python
def typicalSampling(group):
    name = group.name
    frac = 0.008
    return group.sample(frac=frac)

result = major.groupby(
    'klabel', group_keys=False
).apply(typicalSampling)
```

```python
result['Label'].value_counts()
```

```python
result
```

```python

result = result.drop(['klabel'],axis=1)
result = pd.concat([result, minor], ignore_index=True)
```

```python
result.to_csv('/content/CICIDS2017_sample.csv')
```
